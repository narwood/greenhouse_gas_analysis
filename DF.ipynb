```python
import pandas as pd
import numpy as np

# GIS

# Country, Industry, Gas_Type

```

```python
df = pd.read_csv("Annual_Greenhouse_Gas_(GHG)_Air_Emissions_Accounts.csv")
df_use = df[['Country', 'Industry', 'Gas_Type', 'F2010', 'F2011','F2012', 'F2013', 'F2014', 'F2015', 'F2016', 'F2017', 'F2018', 'F2019', 'F2020', 'F2021']]
df_use
```

```python
array = df_use.to_numpy()

# 0=Country, 1=Industry, 2=Gas_Type, 3=F2010, ... , 14=F2021

countries = df_use.Country.unique()
industries = df_use.Industry.unique()
gas_types = df_use.Gas_Type.unique()

countries_totals = {}
for country in countries:
    countries_totals[country] = np.zeros(12, dtype=int)
    
industries_totals = {}
for industry in industries:
    industries_totals[industry] = np.zeros(12, dtype=int)
    
gastypes_totals = {}
for gas_type in gas_types:
    gastypes_totals[gas_type] = np.zeros(12, dtype=int)

for obs in array:
    obs_total = np.array(obs[3:])
    obs_country = obs[0]
    obs_industry = obs[1]
    obs_gastype = obs[2]
    countries_totals[obs_country] = np.add(countries_totals[obs_country], obs_total)
    industries_totals[obs_industry] = np.add(industries_totals[obs_industry], obs_total)
    gastypes_totals[obs_gastype] = np.add(gastypes_totals[obs_gastype], obs_total)

    
print(countries_totals)
print()
print(industries_totals)
print()
print(gastypes_totals)
print()
```
