```python
import pandas as pd
import numpy as np

# GIS

# Country, Industry, Gas_Type

```

```python
df = pd.read_csv("Annual_Greenhouse_Gas_(GHG)_Air_Emissions_Accounts.csv")
df_use = df[['Country', 'Industry', 'Gas_Type', 'F2010', 'F2011','F2012', 'F2013', 'F2014', 'F2015', 'F2016', 'F2017', 'F2018', 'F2019', 'F2020', 'F2021']]
df_use
```

```python
array = df_use.to_numpy()

# 0=Country, 1=Industry, 2=Gas_Type, 3=F2010, ... , 14=F2021

countries = df_use.Country.unique()
industries = df_use.Industry.unique()
gas_types = df_use.Gas_Type.unique()

countries_totals = dict(zip(countries, np.zeros(len(countries), dtype=int)))
industries_totals = dict(zip(industries, np.zeros(len(industries), dtype=int)))
gastypes_totals = dict(zip(gas_types, np.zeros(len(gas_types), dtype=int)))

for obs in array:
    obs_total = sum(obs[3:])
    obs_country = obs[0]
    obs_industry = obs[1]
    obs_gastype = obs[2]
    countries_totals[obs_country] += obs_total
    industries_totals[obs_industry] += obs_total
    gastypes_totals[obs_gastype] += obs_total
    
print(countries_totals)
print()
print(industries_totals)
print()
print(gastypes_totals)
print()
```
